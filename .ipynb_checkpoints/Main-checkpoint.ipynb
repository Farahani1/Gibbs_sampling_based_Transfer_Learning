## DATA PERPRATION
In this section MNIST and MNIST-M dataset are loaded and preprocessed. Also, an RBM which is tarind by MNIST-M data is loaded, and it is used to cast target instances into a new domain close to the source.




```python
# DATA ADRESSES:
# Both MNIST and MNIST-M datasets are provided as pickle files with protocl2

mnist_addr = "./dataset/mnist_data_prtcl2.pkl"
mnistm_addr = "./dataset/mnistm_data_prtcl2.pkl"
rbm_addr = './models/model2'
```

```python
import numpy as np
import pickle
from utils import rgb2gray, display_images, model_trainer, report_accu, plot_confusion_matrix
from pydeep.rbm.sampler import GibbsSampler
from pydeep.misc.io import load_object
%tensorflow_version 1.x


invert = True

# You may also want to transfer source instances with an step of Gibbs sampling. 
sampling_source = False


# Loding mnist and mnist-m row datasets
data = pickle.load( open(mnist_addr, "rb" ))
datam = pickle.load( open( mnistm_addr, "rb" ))

# Collecting test and train data in the source domain
source_train = data['mnist_train']
source_test = data['mnist_test']

# Adding iverted images to source data
if invert:
  source_train = np.concatenate((source_train, 1-source_train), axis=0)
  source_test = np.concatenate((source_test, 1-source_test), axis=0)

# Turn mnist-m_train and test images into graysacle, resahpe and normalize it
target_train = rgb2gray(datam['train'])
target_train = target_train.reshape([target_train.shape[0], 784])
target_train = (target_train-target_train.min())/target_train.max()
initil_states_for_sampling = target_train

# Transfer target instance to the a new domain
rbm = load_object(rbm_addr, info=True)
sampler = GibbsSampler(rbm)
geibbs_steps =1 # number of Gibbs sampling setps
print 'sampling mnistm train'
sampled_target_train = sampler.sample(vis_states=initil_states_for_sampling, k=geibbs_steps, betas=None, ret_states=False) 

# Transfer source instance to the new domain.
# (you need to keep in mind that this new domain is very close to the source.)
if sampling_source:
  print 'sampling mnist train'
  source_train = sampler.sample(vis_states=source_train, k=k, betas=None, ret_states=False)


# Reshaping the dataset
source_train = source_train.reshape(source_train.shape[0], 28, 28)*255
source_test = source_test.reshape(source_test.shape[0], 28, 28)*255
target_train = target_train.reshape(target_train.shape[0], 28, 28)*255
sampled_target_train = sampled_target_train.reshape(sampled_target_train.shape[0], 28, 28)*255

pixel_mean = np.vstack([source_train, sampled_target_train]).mean((0, 1, 2))

if invert:
  data['mnist_train_labels'] = np.concatenate((data['mnist_train_labels'], data['mnist_train_labels']), axis=0)
  data['mnist_test_labels'] = np.concatenate((data['mnist_test_labels'],data['mnist_test_labels']), axis=0)
```

**More illustration about *Temperture* parameter**

The box below provides an illustration for the temperture concept, you can see that how increaseing or decreasing the can effect the final probability distribution.






```python
import numpy as np

def proabilities(e0, e1, T):
  z = np.exp(e0/T) + np.exp(e1/T)  # Partition function
  print "\nIf the temperture is ", T, " :"
  print "P(1) = ", np.exp(e0/T) / z
  print "P(2) = ", np.exp(e1/T) / z ,

# The value of of negetive energys (goodness):
e0 = 5
e1 = 4

proabilities(e0, e1, T=1000)
proabilities(e0, e1, T=1)
proabilities(e0, e1, T=0.01)
```
